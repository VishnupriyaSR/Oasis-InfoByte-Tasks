# 4. Data Visualization

# 1. Distribution Plot of Unemployment Rate

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x=' Estimated Unemployment Rate (%)', kde=True)
plt.title('Distribution of Unemployment Rate')
plt.xlabel('Unemployment Rate (%)')
plt.show()
```

# 2.Scatter Plot of Unemployment Rate vs. Labour Participation Rate

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x=' Estimated Labour Participation Rate (%)', y=' Estimated Unemployment Rate (%)', hue='Region')
plt.title('Unemployment Rate vs. Labour Participation Rate')
plt.xlabel('Labour Participation Rate (%)')
plt.ylabel('Unemployment Rate (%)')
plt.show()
```

# 3.Bar Plot of Unemployment Rate by Region

```python
# Plot
plt.figure(figsize=(12, 8))
sns.barplot(data=df, x='Region', y=' Estimated Unemployment Rate (%)')
plt.title('Unemployment Rate by Region')
plt.xlabel('Region')
plt.ylabel('Unemployment Rate (%)')
plt.xticks(rotation=90)
plt.show()
```

# From the above plot we observe that Tripura,Haryana and Jharkand have the highest Unemployment Rate and Meghalaya have least Unemployment Rate

# 4.Line Plot of Unemployment Rate Over Time

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x=' Date', y=' Estimated Unemployment Rate (%)')
plt.title('Unemployment Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.show()
```

# In above analysis of the unemployment data, we observed a significant spike in the unemployment rate during March to May 2020. During this period, the unemployment rate increased to 24.88, which is notably higher than the average rate observed in the surrounding months. This increase may be indicative of underlying economic challenges or other factors(e.g.,due to corona) affecting the job market during that time.This sharp increase contrasts with the generally stable rates observed earlier in the year, suggesting a possible economic disruption or other significant factors affecting employment during this period.

# 5. Bar Plot for Distribution of Unemployment

```python
plt.figure(figsize=(10,5))
sns.barplot(data=df, x='Area', y=' Estimated Unemployment Rate (%)')
plt.title('Unemployment Rate by Area')
plt.xlabel('Region')
plt.ylabel('Unemployment Rate (%)')
plt.show()
```

# From the above plot it is clear that Unemployment Rate is higher in Urban Areas than Rural Area

# 6.Box Plot

```python
def plot(df,column):
    plt.figure(figsize=(15,5))
    plt.subplot(1,2,1)
    sns.boxplot(data=df,x=column)
    plt.title(f'Box Plot for {column}')

    plt.subplot(1,2,2)
    sns.histplot(data=df,x=column,kde=True,bins=50)
    plt.title(f'Distribution Plot for {column}')

```

```python
for i in [' Estimated Unemployment Rate (%)', ' Estimated Employed', ' Estimated Labour Participation Rate (%)']:
    plot(df,i)
```
